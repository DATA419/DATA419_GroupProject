```python
artist_names=[
'FKAtwigs',
'andygrammer',
'c_woodzofficial',
'beforeyouexit',
'rrr001222',
'CAVETOWN',
'clairo',
'oneruel',
'Olivertree',
'TessaViolet',
'alinabaraz',
'CigsAfterSexx',
'KygoMusic',
'ninanesbitt',
'cleanbandit',
'chaseatlantic',
'tatemcrae',
'hellohonne',
'JorjaSmith',
'kimpetras']
```

```python
# Twitter API v2 OAuth 2.0 Bearer Token
client = tweepy.Client(bearer_token)
```

```python
# get user id+name objects from artist names
artist_user_objs = []
for i in artist_names:
  a = client.get_user(username = i)
  artist_user_objs.append(a.data)
```

```python
artist_user_objs
```

```python
id_number = []
id_name = []
id_username = []

for i in artist_user_objs:
    number = i['id']
    name = i['name']
    username = i["username"]
    id_number.append(number)
    id_name.append(name)
    id_username.append(username)

celeb_id_data = {
    'celeb_id_number': id_number,
    'celeb_id_name': id_name,
    "celeb_id_username": id_username
}

celeb_id_df = pd.DataFrame(celeb_id_data)
```

```python
celeb_id_df
```

```python
# empty object for list of lists
tweets_list = []
```

```python
# will get rate limit, wait 15-20 mins and run this cell again

# range will start (0,19) for celebs 0 to 18
# after rate limit, should be 15 celebs done, for 0 to 14
# range will then be (15,19) for celebs 15 to 18

for i in range(len(tweets_list),len(celeb_id_df["celeb_id_number"])):
    id_num = celeb_id_df["celeb_id_number"][i]
    users_tweets = client.get_users_tweets(id = id_num, max_results = 50)
    tweets_list.append(users_tweets)
```

```python
#[0-18] for accounts
#[0] for response
#[0-99] for tweet objects
#["id" or "text"]
#tweets_list[8][0][88]["text"]
```

```python
# gets tweet ids from objects, then tweet ids into dataframe 
celeb_tweets_vec = []

for i in range(0,len(tweets_list)):
    sublist =[]
    for j in tweets_list[i][0]:
        sublist.append(j["id"])
    celeb_tweets_vec.append(sublist)

celeb_tweets_data = {
'celeb_id_number': id_number,
'tweet_id_number' : celeb_tweets_vec
}

# one row per celeb, with list of tweet IDs in 2nd column
celeb_tweets_df = pd.DataFrame(celeb_tweets_data)
# long format, one row per tweet ID
celeb_tweets_df_long  = celeb_tweets_df.explode("tweet_id_number", ignore_index=True)
```

```python
all = pd.merge(celeb_tweets_df, celeb_id_df, on = 'celeb_id_number')
```
